```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# dataset from https://www.kaggle.com/andrewmvd/heart-failure-clinical-data
df = pd.read_csv('csv/heart_failure_clinical_records_dataset.csv')
```

```python
df.head(10)
```

```python
df1 = df[['age','diabetes','sex','smoking','DEATH_EVENT']]
```

```python
# Gender percentage of patients with heart failure

heart_failure_sex = df1.groupby('sex').count()['age']
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
plt.pie(heart_failure_sex,labels=['Female','Male'],autopct='%0.2f%%')
plt.title('Gender percentage of patients with heart failure')
plt.axis('equal')
plt.show()
```

```python
# Percentage of patients with diabetes

heart_failure_diabetes = df1.groupby('diabetes').count()['age']
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
plt.pie(heart_failure_diabetes,labels=['Without diabetes','With diabetes'],autopct='%0.2f%%')
plt.title('Percentage of patients with diabetes')
plt.axis('equal')
plt.show()
```

```python
# Patient is a smoker

heart_failure_smoking = df1.groupby('smoking').count()['age']
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
plt.pie(heart_failure_smoking,labels=['No','Yes'],autopct='%0.2f%%')
plt.title('Patient is a smoker')
plt.axis('equal')
plt.show()
```

```python
# Age of Patients

fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
sns.distplot(df1['age'],ax=ax)
plt.title('Age of Patients')
plt.show()
```
