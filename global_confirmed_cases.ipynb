```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# dataset from https://www.kaggle.com/antgoldbloom/covid19-data-from-john-hopkins-university
df = pd.read_csv('csv/global_confirmed_cases.csv')
```

```python
df.head(10)
```

```python
df1 = df.drop([0]).rename(columns={'Country/Region':'Date'})
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df2 = df1.set_index('Date').applymap(lambda x: int(float(x)))
```

```python
df2.columns.values
```

```python
# aggregate countries with state data

australia_sub = ['Australia.'+str(i) for i in range(1,8)]
canada_sub = ['Canada.'+str(i) for i in range(1,16)]
china_sub = ['China.'+str(i) for i in range(1,33)]
denmark_sub = ['Denmark.'+str(i) for i in range(1,3)]
france_sub = ['France.'+str(i) for i in range(1,11)]
netherlands_sub = ['Netherlands.'+str(i) for i in range(1,5)]
uk_sub = ['United Kingdom.'+str(i) for i in range(1,11)]
df2['Australia'] = df2[['Australia']+australia_sub].sum(axis=1)
df2['Canada'] = df2[['Canada']+canada_sub].sum(axis=1)
df2['China'] = df2[['China']+china_sub].sum(axis=1)
df2['Denmark'] = df2[['Denmark']+denmark_sub].sum(axis=1)
df2['France'] = df2[['France']+france_sub].sum(axis=1)
df2['Netherlands'] = df2[['Netherlands']+netherlands_sub].sum(axis=1)
df2['United Kingdom'] = df2[['United Kingdom']+uk_sub].sum(axis=1)
```

```python
df3 = df2.drop(australia_sub+canada_sub+china_sub+denmark_sub+france_sub+netherlands_sub+uk_sub,axis=1)
```

```python
df3.head(7)
```

```python
# Top 10 Countries with most confirmed cases

top10 = df3.sum().sort_values(ascending=False).head(10).sort_values()
```

```python
top10
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
ax.barh(top10.index,top10/1000000)
ax.set_title('Top 10 Countries with Most Confirmed Cases')
ax.set_xlabel('Million')
plt.show()
```

```python
# Weekly confirmed cases from top 3 countries with highest cases

global_weekly_confirmed = df3[['US','India','Brazil']].resample('W').sum()[:'2020-12-07']
```

```python
global_weekly_confirmed.head()
```

```python
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
ax.plot(global_weekly_confirmed.index,global_weekly_confirmed,lw=3)
ax.legend(global_weekly_confirmed.columns)
ax.set_title('Weekly confirmed cases from top 3 countries with highest cases')
ax.set_xlabel('Date')
ax.grid()
plt.show()
```

```python
# Philippines weekly confirmed cases

philippines_weekly = df3['Philippines'].resample('W').sum()[:'2020-12-07']
```

```python
philippines_weekly.head()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1.4,1.4])
ax.plot(philippines_weekly.index,philippines_weekly,lw=3)
ax.legend(['Philippines'])
ax.set_title('Philippines weekly confirmed cases')
ax.set_xlabel('Date')
ax.grid()
plt.show()
```
